```python
import pandas as pd
```

```python
df = pd.read_csv("data/resume_data.csv", encoding="utf-8-sig")
```

```python
df.columns = df.columns.str.replace("\ufeff", "", regex=False).str.strip()
df.columns.tolist()
```

```python
df = df[["skills_required", "job_position_name"]]
df.head()
```

```python
df = df.rename(columns={"skills_required": "skills"})
df.head()
```

```python
df = df.dropna(subset=["skills", "job_position_name"])
df.shape
```

```python
df.head()
```

```python
import re

def clean_text(text):
    text = str(text)
    text = text.lower()

    # remove URLs
    text = re.sub(r"http\S+", "", text)

    # keep +, #, ., / (important for skills)
    text = re.sub(r"[^a-zA-Z0-9+#./ ]", " ", text)

    # normalize spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df["skills"] = df["skills"].apply(clean_text)
df.head()
```

```python
df = df[df["skills"].str.strip() != ""]
df.shape
```

```python
df.head()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.head()
```

```python
df.to_csv("clean_skills_dataset.csv", index=False)
print("Clean dataset saved successfully!")
```
