```python
import random
import pandas as pd

job_roles = {
    "Data Scientist": ["python", "pandas", "numpy", "machine learning", "statistics", "data visualization", "sklearn"],
    "Data Analyst": ["excel", "sql", "python", "data cleaning", "tableau", "powerbi"],
    "Machine Learning Engineer": ["python", "tensorflow", "pytorch", "deep learning", "mlops", "model deployment"],
    "Frontend Developer": ["html", "css", "javascript", "react", "redux", "responsive design"],
    "Backend Developer": ["python", "django", "flask", "api development", "sql", "database"],
    "Full Stack Developer": ["html", "css", "javascript", "react", "nodejs", "mongodb"],
    "DevOps Engineer": ["docker", "kubernetes", "ci cd", "aws", "linux", "monitoring"],
    "Cloud Engineer": ["aws", "azure", "gcp", "networking", "security", "cloud architecture"],
    "Software Engineer": ["java", "c++", "data structures", "algorithms", "system design"],
    "AI Engineer": ["python", "nlp", "computer vision", "deep learning", "tensorflow"],
    "Cybersecurity Analyst": ["network security", "encryption", "ethical hacking", "vulnerability assessment"],
    "Database Administrator": ["sql", "database management", "backup", "recovery", "performance tuning"],
    "Mobile App Developer": ["java", "kotlin", "android", "ios", "swift"],
    "QA Engineer": ["testing", "automation", "selenium", "debugging", "test cases"],
    "Blockchain Developer": ["solidity", "ethereum", "smart contracts", "web3"],
}

data = []

for _ in range(500):
    role = random.choice(list(job_roles.keys()))
    skills = random.sample(job_roles[role], k=random.randint(4, len(job_roles[role])))
    data.append([role, " ".join(skills)])

df = pd.DataFrame(data, columns=["job_role", "skills"])

df.to_csv("it_jobs_dataset.csv", index=False)

print("CSV with 500 rows created!")
```
